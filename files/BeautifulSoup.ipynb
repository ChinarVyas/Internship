# 1) Write a python program to display all the header tags from wikipedia.org and make data frame.

```python
# URL of the webpage to scrape
url = "https://en.wikipedia.org/wiki/Main_Page"

response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.text, "html.parser")

    # Find and print all header tags (h1, h2, h3, etc.)
    header_tags = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])

    for tag in header_tags:
        print(tag.text.strip())
else:
    print("Failed to retrieve the web page. Status code:", response.status_code)

```

# 3) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data frame-
a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
b) Top 10 ODI Batsmen along with the records of their team andrating.
c) Top 10 ODI bowlers along with the records of their team andrating.

```python
#Top 10 ODI Teams
def scrape_top_teams():
    url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    table = soup.find('table', {'class': 'table'})
    rows = table.find_all('tr')[1:11]  # Get top 10 rows

    data = []
    for row in rows:
        cols = row.find_all('td')
        team = cols[1].text.strip()
        matches = cols[2].text.strip()
        points = cols[3].text.strip()
        rating = cols[4].text.strip()
        data.append([team, matches, points, rating])

    df = pd.DataFrame(data, columns=['Team', 'Matches', 'Points', 'Rating'])
    return df

df = scrape_top_teams()
print(df)
```

# 5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame-
i) Headline
ii) Time
iii) News Link

```python
url="https://www.cnbc.com/world/?region=world"
response= requests.get(url)
print(response)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
#Headlines
headlines=[]
for i in soup.find_all("a",class_="LatestNews-headline"):
    headlines.append(i.text)

#Time
times=[]
for i in soup.find_all("time",class_="LatestNews-timestamp"):
    times.append(i.text)
#Links 

n_links=[]
for i in soup.find_all("a",class_="LatestNews-headline"):
    n_links.append(i['href'])
```

```python
cnbc ={"Headlines":headlines, "Time":times,"Profile Links":n_links}

CNBC=pd.DataFrame(cnbc)
CNBC
```

# 7) Write a python program to scrape mentioned details from dineout.co.in and make data frame-
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
# URL of the page to scrape
url = "https://www.dineout.co.in/vadodara-restaurants"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.text, "html.parser")

    # Lists to store the scraped data
    restaurant_names = []
    cuisines = []
    locations = []
    image_urls = []

    # Find and extract the relevant information
    restaurant_cards = soup.find_all("div", class_="restnt-card restaurant")

    for card in restaurant_cards:
        # Restaurant Name
        restaurant_name_element = card.find("div", class_="restnt-info cursor")
        restaurant_name = restaurant_name_element.text.strip() if restaurant_name_element else "N/A"
        restaurant_names.append(restaurant_name)

        # Cuisine
        cuisine_element = card.find("span", class_="double-line-ellipsis")
        cuisine = cuisine_element.text.strip() if cuisine_element else "N/A"
        cuisines.append(cuisine)

        # Location
        location_element = card.find("div", class_="restnt-loc ellipsis")
        location = location_element.text.strip() if location_element else "N/A"
        locations.append(location)

        # Image URL
        image_element = card.find("div", class_="img cursor")
        if image_element:
            image_url = image_element.find('img')['data-src']
        else:
            image_url = "N/A"
        image_urls.append(image_url)

    # Create a DataFrame
    data = {
        "Restaurant Name": restaurant_names,
        "Cuisine": cuisines,
        "Location": locations,
        "Image URL": image_urls,
    }
    df = pd.DataFrame(data)

else:
    print("Failed to retrieve the web page. Status code:", response.status_code)
df
```
