1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
delay = 5
url ='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver = webdriver.Chrome()
driver.get(url)
driver.fullscreen_window()
WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CLASS_NAME,'mw-footer')))
```

```python
Rank =[]
Name=[]
Artist=[]
Upload=[]
Views=[]

#Xpath to the table& row

table = driver.find_element(By.XPATH, '//*[@id="mw-content-text"]/div[1]/table[3]')  
rows = table.find_elements(By.TAG_NAME, "tr")

for row in rows:
    cols = row.find_elements(By.TAG_NAME, "td")
    if cols:
        Rank.append(cols[0].text)
        Name.append(cols[1].text)
        Artist.append(cols[2].text)
        Upload.append(cols[4].text)
        Views.append(cols[3].text)

Youtube_video={"Rank":Rank,"Name": Name,"Artist":Artist, "Date_Uploaded":Upload,"Total_Views":Views}

Youtube_videos=pd.DataFrame(Youtube_video)
driver.quit()
```

2. Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Series
B) Place
C) Date
D) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
url='https://www.bcci.tv/'

driver = webdriver.Chrome()
driver.get(url)
driver.maximize_window()

# Wait for the page to load
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '/html/body/div[8]/footer/div/div/div/div[2]/p')))

# Navigate to the International Fixtures page
fixtures = driver.find_element(By.XPATH, '/html/body/header/div[3]/div[1]/ul/div[1]/a[2]')
fixtures.click()

try:
    time.sleep(8)
    # Click the dropdown menu to open it
    dropdown_menu = driver.find_element(By.XPATH, '/html/body/section/div/div/div/div/div/div[2]/div[1]/div[2]/div/div[1]/div/div[1]') 
    dropdown_menu.click()
    time.sleep(3)  # wait for the dropdown menu to open

    # Click the option you want to select
    option = driver.find_element(By.XPATH, '/html/body/section/div/div/div/div/div/div[2]/div[1]/div[2]/div/div[1]/div/div[2]/div[11]')
    option.click()
except NoSuchElementException:
    print("Dropdown menu or option not found.")
    
series_list = []
place_list = []
date_list = []
time_list = []

matches = driver.find_elements(By.XPATH,'/html/body/section/div/div/div/div/div/div[2]/div[2]/div[1]/div[1]/div')

for match in matches:
    # Extract the details
    series = match.find_elements(By.XPATH, '//h5[@class="match-tournament-name ng-binding"]')
    place = match.find_elements(By.XPATH, '//span[@class="ng-binding"]')
    date = match.find_elements(By.XPATH, '//div[@class="match-dates ng-binding"]')
    time = match.find_elements(By.XPATH, '//div[@class="match-time no-margin ng-binding"]')

for i in series:
    series_list.append(i.text)
for i in place:
    place_list.append(i.text)
for i in date:
    date_list.append(i.text)
for i in time:
    time_list.append(i.text)

# Create a DataFrame from the lists
df = pd.DataFrame({
    'Series': series_list,
    'Place': place_list,
    'Date': date_list,
    'Time': time_list
})

# Save the DataFrame to a CSV file
#df.to_csv('bcci_fixtures.csv', index=False)

driver.quit()
```

3. Scrape the details of State-wise GDP of India from statisticstime.com.
Url = http://statisticstimes.com/
You have to find following details: A) Rank
B) State
C) GSDP(18-19)- at current prices
D) GSDP(19-20)- at current prices
E) Share(18-19)
F) GDP($ billion)
Note: - From statisticstimes home page you have to reach to economy page through code.

```python
URL ='https://www.statisticstimes.com/economy/india/indian-states-gdp.php'


# Create a new instance of the Firefox driver
driver = webdriver.Chrome()

# Go to the website
driver.get("http://statisticstimes.com/")

# Click on the 'Economy' dropdown menu
economy = driver.find_element(By.XPATH,'/html/body/div[2]/div[1]/div[2]/div[2]/button')
economy.click()

# Click on the 'India' option
india = driver.find_element(By.XPATH,'/html/body/div[2]/div[1]/div[2]/div[2]/div/a[3]')
india.click()

# Get the page source and create a BeautifulSoup object
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find the table and table data
table = soup.find_all('table')[0]
rows = table.find_all('tr')

# Create a list to store the data
all_data = []

for row in rows:
    cols = row.find_all('td')
    cols = [col.text.strip() for col in cols]
    all_data.append(cols)

# Create a DataFrame from the list of data
df = pd.DataFrame(all_data, columns=["Rank", "State", "GSDP(18-19)", "GSDP(19-20)", "Share(18-19)", "GDP($ billion)"])

# Print the DataFrame
print(df)

# Close the browser
driver.quit()
```

# Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used

```python
import time
# connecting to the web driver
driver=webdriver.Chrome()

# getting the webpage of mentioned url
url = ("https://github.com/")
driver.get(url)
driver.maximize_window()

time.sleep(4)
# getting explore button and clicking on it
explore = driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button").click()

time.sleep(4)
# selecting trending option
trend_url = driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a").click()
#driver.get(urls)
```

```python
# creating empty list
URLs = []
repository_title = []
Description = []
Contributors = []
Language = []
lang = []

# fetching urls for each repository
repository = driver.find_elements(By.XPATH,'//a[@class="Link"]')
for i in repository:
    URLs.append(i.get_attribute("href"))
    
# scraping Repository title data
title = driver.find_elements(By.XPATH,'//a[@class="Link"]')
for i in title:
    repository_title.append(i.text)
    
# scraping data from all repository page
for i in URLs:
    driver.get(i)
    time.sleep(5)
    
    # scraping Repository Description data 
    try:
        desc = driver.find_element(By.XPATH,"//p[@class='f4 mt-3']")
        Description.append(desc.text)
    except NoSuchElementException:
        Description.append('-')
        
        
    # scraping Contributors Count data
    try:
        contributor = driver.find_element(By.XPATH,"//*[contains(text(),'    Contributors ')]")
        Contributors.append(contributor.text.replace('Contributors',''))
    except NoSuchElementException:
        Contributors.append('-')
    
    
    # scraping Languages used data
    try:
        for i in driver.find_elements(By.XPATH,"//ul[@class= 'list-style-none']//li//span[1]"):
            lang.append(i.text)
        Language.append(lang)
    except NoSuchElementException:
        Language.append('-')
        
        
# Data Framing
Github = pd.DataFrame({})
Github['Repository Title'] = repository_title
Github['Repository Description'] = Description
Github['Contributors Count'] = Contributors
Github['Language Used'] = Language
Github
```

#6Scrape the details of Highest selling novels.
A) Book name
B) Author name
C) Volumes sold
D) Publisher
E) Genre
Url - https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare

```python
# Initialize the driver
driver = webdriver.Chrome()
url = ("https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare/")
driver.get(url)
driver.maximize_window()
time.sleep(3)

# creating empty lists
Book_name = []
Author_name = []
Volumes_sold = []
Publisher = []
Genre = []

# scraping book names data
for i in driver.find_elements(By.XPATH,"//tbody//tr//td[2]"):
    Book_name.append(i.text)

# scraping author names data
for i in driver.find_elements(By.XPATH,"//tbody//tr//td[3]"):
    Author_name.append(i.text if i.text != '0' else '-')

time.sleep(3)

# scraping data of volumes sold
for i in driver.find_elements(By.XPATH,"//tbody//tr//td[4]"):
    Volumes_sold.append(i.text)

# scraping data of publisher names
for i in driver.find_elements(By.XPATH,"//tbody//tr//td[5]"):
    Publisher.append(i.text)

# scraping  data of genre
for i in driver.find_elements(By.XPATH,"//tbody//tr//td[6]"):
    Genre.append(i.text)

# creating dataframe for scraped data
Novels = pd.DataFrame({
    'Book Name': Book_name,
    'Author': Author_name,
    'Volume sold': Volumes_sold,
    'Publisher': Publisher,
    'Genre': Genre
})

driver.quit()

Novels
```

7. Scrape the details most watched tv series of all time from imdb.com.
Url = https://www.imdb.com/list/ls095964455/ You have to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes

```python
# Initialize the driver
driver = webdriver.Chrome()

# Navigate to the webpage
driver.get("https://www.imdb.com/list/ls095964455/")

# creating empty lists
Name = []
Year_span = []
Genre = []
Run_time = []
Ratings = []
Votes = []

# scraping data
for show in driver.find_elements(By.XPATH,"//div[@class='lister-item mode-detail']"):
    Name.append(show.find_element(By.XPATH,".//h3[@class='lister-item-header']//a").text)
    Year_span.append(show.find_element(By.XPATH,".//h3[@class='lister-item-header']//span[2]").text)
    Genre.append(show.find_element(By.XPATH,".//p[@class='text-muted text-small']//span[5]").text)
    Run_time.append(show.find_element(By.XPATH,".//p[@class='text-muted text-small']//span[3]").text)
    Ratings.append(show.find_element(By.XPATH,".//div[@class='ipl-rating-star small']//span[2]").text)
    Votes.append(show.find_element(By.XPATH,".//p[4]//span[2]").text)

# creating dataframe for scraped data
TV_Shows = pd.DataFrame({
    'Name': Name,
    'Year Span': Year_span,
    'Genre': Genre,
    'Run Time': Run_time,
    'Ratings': Ratings,
    'Votes': Votes
})
# Close the driver
driver.quit()
```
