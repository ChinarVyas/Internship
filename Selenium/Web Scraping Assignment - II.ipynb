# Q1: Write a python program to scrape data for ‚ÄúData Analyst‚Äù Job position in ‚ÄúBangalore‚Äù location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

2. Enter ‚ÄúData Analyst‚Äù in ‚ÄúJob title, Skills‚Äù field and enter ‚ÄúBangalore‚Äù in ‚Äúenter the location‚Äù field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.

```python
url ="https://www.shine.com/"

#creating chrome instance
driver=webdriver.Chrome()

#1)accessing the website

driver.get(url)
#maximizing the window
time.sleep(2)
driver.maximize_window()

#the notification takes approx 10-12 secs to pop up hence time set to 14
time.sleep(10)

#selecting the(x) to avoid further notifications

notification=driver.find_element(By.XPATH,"/html/body/div/div[2]/div/div/button")
notification.click()
time.sleep(7)

#moving on to fill details like Job Title & Location and clicking on search button
#using time to allow the page to load
click=driver.find_element(By.XPATH,"/html/body/div/header[1]/div[3]/div/div/div[1]/div/span/i")
click.click()

time.sleep(3)
designation= driver.find_element(By.XPATH,'/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[1]/div/input')
designation.send_keys("Data Analyst")
time.sleep(2)

location=driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input")
location.send_keys("Bangalore")
time.sleep(2)
search=driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[2]/div/button")
search.click()

time.sleep(15)
notification_1 = driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div/div/button")
notification_1.click()
```

```python
#checking the size of the data extrated:-
print(len(job_title),len(job_location),len(company_name))
```

```python
#creating the dataframe
jobs=({"Title":job_title,"Location":job_location,"Company Name":company_name})
vacancies=pd.DataFrame(jobs)
vacancies
```

# Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
6. Brand
7. Product Description
8. Price
The attributes which you have to scrape is ticked marked in the below image.

```python
url = "https://www.flipkart.com/"

driver=webdriver.Chrome()

driver.get(url)
#maximising the window
driver.maximize_window()
time.sleep(1)
#using time above to give time to the page to load then click on the (X).

notification=driver.find_element(By.XPATH,"/html/body/div[3]/div/span")
notification.click()
time.sleep(1)

#input sunglasses in search field
product = driver.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/header/div[1]/div[2]/form/div/div/input")
product.send_keys("Sunglasses")

#click on search
productsearch = driver.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/header/div[1]/div[2]/form/div/div/input")
productsearch.submit()

brand=[]
product_description=[]
price =[]
discount=[]
```

```python
start=0
end=3
time.sleep(2)
for page in range(start,end):
    G_brand =driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in G_brand:
        if len(brand)>=100:
            break
        G_b=i.text
        brand.append(G_b)

#product description    
    prod_des=driver.find_elements(By.CLASS_NAME,"IRpwTa")
    for i in prod_des:
        if len(product_description)>=100:
            break
        prod=i.text
        product_description.append(prod)

#price
    rate=driver.find_elements(By.CLASS_NAME,"_30jeq3")
    for i in rate:
        if len(price)>=100:
            break
        prices=i.text
        price.append(prices)

#Discount
    discounts=driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]/span')
    for i in discounts:
        if len(discount)>=100:
            break
        dis=i.text
        discount.append(dis)
    next_button=driver.find_element(By.XPATH,'//span[text()="Next"]')
    next_button.click()
    time.sleep(3)
```

```python
print(len(brand),len(product_description),len(price),len(discount))
```

```python
sunglass={"Brands":brand,"Products_Description":product_description,"Price":price,"Discounts":discount}
```

```python
sunglasses =pd.DataFrame(sunglass)
sunglasses
```

# Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on Top Quotes
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
url='https://www.azquotes.com/'
driver=webdriver.Chrome()
driver.get(url)
time.sleep(2)
driver.maximize_window()
time.sleep(2)
click=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
click.click()
```

```python
#creating empty shells to save the data
Quote=[]
Author=[]
TQ=[]

#setting the range to iterate the extraction
start = 0
end= 10
for page in range(start,end):
    qte=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in qte:
        if len(Quote)>=1000:
            break
        qt=(i.text)
        Quote.append(qt)
    auth=driver.find_elements(By.XPATH,'//div[@class="author"]/a')
    for i in auth:
        if len(Author)>=1000:
            break
        at=i.text
        Author.append(at)
    toq=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in toq:
        if len(TQ)>=1000:
            break
        at=i.text
        TQ.append(at)
    next_step=driver.find_element(By.XPATH, '//div[@class="pager"]/li[12]/a')
    next_step.click()
    time.sleep(3)
```

```python
print(len(Quote),len(Author),len(TQ))
```

```python
gyan={"Quote":Quote,"Author":Author,"Type of Quotes": TQ}
Quotes=pd.DataFrame(gyan)
Quotes
```

# Q9: Write a python program to display list of respected former Prime Ministers of India(i.e. Name, Born-Dead, Term of office, Remarks) from https://www.jagranjosh.com/.
This task will be done in following steps:
1. First get the webpagehttps://www.jagranjosh.com/
2. Then You have to click on the GK option
3. Then click on the List of all Prime Ministers of India
4. Then scrap the mentioned data and make theDataFrame.

```python
url='https://www.jagranjosh.com/'
driver=webdriver.Chrome()
driver.get(url)
time.sleep(2)
driver.maximize_window()
time.sleep(2)
search=driver.find_element(By.LINK_TEXT,"GK")
search.click()
time.sleep(3)
access=driver.find_element(By.LINK_TEXT,"List of all Prime Ministers of India")
access.click()
```

```python
#creating a shell
data=[]

# Find table
table = driver.find_element(By.TAG_NAME, 'table')

# Get all rows
rows = table.find_elements(By.TAG_NAME, 'tr')

# Iterate over each row
for row in rows:
    # Get all columns
    cols = row.find_elements(By.TAG_NAME, 'td')
    # Extract and print data
    if len(cols) > 0:
        data.append({
            'Name': cols[0].text,
            'Born - Death': cols[1].text,
            'Term of Office': cols[2].text,
            'Remarks': cols[3].text
        })
# Close driver
driver.quit()
```

```python
df=pd.DataFrame(data)
df
```
